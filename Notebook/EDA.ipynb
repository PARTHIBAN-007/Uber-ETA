```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Importing the Data
df = pd.read_csv(r"D:\Projects\Uber\Data\data_cleaned.csv")
```

```python
cols = ['Delivery_person_Age','Delivery_person_Ratings','Weather_conditions','Road_traffic_density','Multiple_deliveries','Festival','City_type']

plt.figure(figsize=(20,15))
i =1
for col in cols:
    plt.subplot(5,2,i)
    i+=1
    sns.countplot(data =df,x =col,order=df[col].value_counts().sort_index().index)
    plt.xlabel(col)
    plt.ylabel("No of Orders")

plt.tight_layout()
plt.show()    
```

# Inference
1. Customers Provide Good Feedback About the Services
2. Ordering Food Items is Same over all Weather Conditions
3. Traffic is A common issue it's not a occassional issue
4. Mostly One Delivery IS Happening at time
5. Orders are very low during Festival Times
6. Most of thr orders are from Metropolitan cities followed by urban and poor ordering in semi-urban
